```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('ggplot')
plt.rcParams["figure.figsize"] = (10,6)
```

```python
df = pd.read_csv('/Users/olivianystrom/Documents/Springboard/bird_flu_dataset.csv')

# Avian influenza or "Bird Flu" is a contagious and often fatal viral disease of birds. 
# Wild birds, particularly wild migratory water birds are considered to be the main reservoir 
# of avian influenza viruses.

# There is a constant risk of avian influenza being introduced into Ireland from 
# wild birds particularly from November onwards each year as this is when migratory birds 
# arrive and congregate on wetlands, mixing with resident species.

# The attached file is a data set of the locations of bird species captured in Ireland from 
# 1980 to 2020 and wild birds that are targeted for the H5N1 strain of avian flu.
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df.dtypes
df[df['Year'] ==2003]
```

```python

agg_year = df.groupby(df['Date'].dt.year).agg({'target_H5_HPAI': sum})
agg_species = df.groupby(df['Common_Name']).agg({'target_H5_HPAI': 'sum', 'Common_Name': 'count'})
agg_species.rename(columns = {'Common_Name':'Individuals_Tested', 'target_H5_HPAI':'Tested_Positive'}, inplace = True)
tested_positive = agg_species[agg_species['Tested_Positive'] > 0]
tested_positive

# this shows how many of each species tested positive and the number of birds tested. Every 
# species that had a bird test positive had ALL tested birds test positive. To me this raises 
# concerns regarding the quality of the data. I will proceed with treating the data as accurate.
```

```python
# taking a look at all the species that tested positive and showing visually how many per species
# were identified as infected. Based on this list of species, this seems to be a waterbird's 
# disease. 
tested_positive2 = tested_positive.reset_index()
tested_positive2 = tested_positive2.sort_values(by='Tested_Positive')
plt.barh(tested_positive2['Common_Name'], tested_positive2['Tested_Positive'])
plt.show()
```

```python
totals_df = df.groupby(['target_H5_HPAI'])['target_H5_HPAI'].count()
```

```python
# Pie chart showing the proportion of infected vs. non-infected individuals
plt.pie(totals_df, labels = ['Not Infected', 'Infected'])
plt.show()
```

```python
#threw the infected/not-infected data into a stacked bar chart by year to see how the populations
# have changed over time. 
totals_yearly = df.groupby(['Year']).aggregate({'target_H5_HPAI': ['sum', 'count']})
                                   
df_yearly = pd.DataFrame(totals_yearly)
# totals_yearly.plot(kind = 'bar', stacked = True) 
# tested_positive2['Common_Name'], tested_positive2['Tested_Positive'])
df_yearly['Not_Infected'] = df_yearly[('target_H5_HPAI', 'count')] - df_yearly[('target_H5_HPAI',   'sum')]
df_yearly['Infected'] = df_yearly[('target_H5_HPAI',   'sum')]
df_yearly[[(  'Not_Infected',      ''), (      'Infected',      '')]].plot(kind = 'bar', stacked = True)
plt.show()
# df_yearly.columns
```

```python
# By far the majority of these infections are identified in Dublin County

totals_by_county = df.groupby(['County']).aggregate({'target_H5_HPAI': ['sum', 'count']})
                                   
df_county = pd.DataFrame(totals_by_county)
# totals_yearly.plot(kind = 'bar', stacked = True) 
# tested_positive2['Common_Name'], tested_positive2['Tested_Positive'])
df_county['Not_Infected'] = df_county[('target_H5_HPAI', 'count')] - df_county[('target_H5_HPAI',   'sum')]
df_county['Infected'] = df_county[('target_H5_HPAI',   'sum')]
df_county = df_county[df_county[(      'Infected',      '')] > 0].sort_values((      'Infected',      ''))
plt.barh(df_county.index, df_county[(      'Infected',      '')])
plt.show()
# df_county
```
